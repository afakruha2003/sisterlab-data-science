```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

birds = pd.read_csv("../birds.csv")

# Genel görünüm
print(birds.head())
print(birds.describe())

# Gelişmiş Scatter Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x="MaxBodyMass", y="MaxWingspan", data=birds, hue="Name", legend=False, palette="viridis")
plt.title("Kuşların Ağırlığı vs Kanat Açıklığı", fontsize=16)
plt.xlabel("Ağırlık (gram)")
plt.ylabel("Maksimum Kanat Açıklığı (cm)")
plt.grid(True)

# Outlier etiketle
for i in range(len(birds)):
    x = birds["MaxBodyMass"][i]
    y = birds["MaxWingspan"][i]
    name = birds["Name"][i]
    if y > 500:
        plt.text(x + 50, y, name, fontsize=9, color="red")

plt.show()

# Korelasyon
corr = birds[["MaxBodyMass", "MaxWingspan"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

birds = pd.read_csv("../birds.csv")
plt.figure(figsize=(14, 6))

# Çubuk grafik - kuş isimlerine göre maksimum kanat açıklığı
plt.bar(birds["Name"], birds["MaxWingspan"], color="skyblue")
plt.title("Kuş Türlerine Göre Maksimum Kanat Açıklığı", fontsize=14)
plt.xlabel("Kuş Türü")
plt.ylabel("Maksimum Kanat Açıklığı (cm)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(birds["MaxWingspan"], marker="o", color="green")
plt.title("Sıralı Maksimum Kanat Açıklığı")
plt.xlabel("Index (Kuş)")
plt.ylabel("Maksimum Kanat Açıklığı (cm)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(birds["MaxWingspan"], bins=10, color="orange", edgecolor="black")
plt.title("Kanat Açıklığı Dağılımı")
plt.xlabel("Maksimum Kanat Açıklığı (cm)")
plt.ylabel("Kuş Sayısı")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(y=birds["MaxWingspan"], color="lightblue")
plt.title("Max Wingspan - Boxplot")
plt.ylabel("Maksimum Kanat Açıklığı (cm)")
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(x="MaxBodyMass", y="MaxWingspan", data=birds)
plt.title("Ağırlık ve Kanat Açıklığı İlişkisi")
plt.xlabel("Ağırlık (gram)")
plt.ylabel("Maksimum Kanat Açıklığı (cm)")
plt.grid(True)
plt.show()
```

```python
correlation = birds[["MaxWingspan", "MaxBodyMass"]].corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Korelasyon Matrisi")
plt.show()
```

```python
# Kuş türlerinin oranı
bird_counts = birds["Name"].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(bird_counts, labels=bird_counts.index, autopct='%1.1f%%')
plt.title("Kuş Türlerinin Dağılımı (Oransal)")
plt.show()
```

```python
plt.figure(figsize=(14, 14))
sns.pairplot(birds[["MaxBodyMass", "MaxWingspan"]])
plt.suptitle("Değişkenler Arası İlişki", y=1.02)
plt.show()
```

```python

plt.figure(figsize=(14, 6))
sns.countplot(data=birds, x="Name", order=birds["Name"].value_counts().index, palette="viridis")
plt.title("Kuş Türlerinin Görülme Sıklığı")
plt.xlabel("Kuş Türü")
plt.ylabel("Frekans")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

text = " ".join(birds["Name"])
wordcloud = WordCloud(width=1000, height=400, background_color="white").generate(text)

plt.figure(figsize=(15, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Kuş İsimleri Kelime Bulutu", fontsize=16)
plt.show()
```
